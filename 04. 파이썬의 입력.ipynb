* 데이터를 입력 받을 수 있는 함수
* 데이터를 문자형으로 입력
 

```python
name = input()
print(name)
```

```python
name = input('이름을 입력하세요')
print('%s님 안녕하세요' %name)
```

```python
# split(): 매개변수를 전달된 문자를 기준으로 데이터를 나눠서 저장
name, age, weight = input('이름, 나이, 몸무게를 입력하세요: ').split(' ')# 슬러시를 넣어두면 슬러시로 입력하라
print('{0}님의 나이는 {1} 살이고, 몸무게는 {2}kg 입니다.'.format(name, age, weight))
```

```python
name, age, weight = input('이름, 나이, 몸무게를 입력하세요: ').split('/')# 슬러시를 넣어두면 슬러시로 입력하라
print('{0}님의 나이는 {1} 살이고, 몸무게는 {2}kg 입니다.'.format(name, age, weight))
```

```python
kor, eng, math = input('국어, 영어, 수학점수를 입력하세요').split(' ')
#print('초점 : ', kor+eng+math) #str 인풑은 스트링으로 나오기 때문에 연결이 됨

# int() : 문자열을 숫자(정수) 타입으로 변환
tot = int(kor) + int(eng) +int(math) # 숫자로 변환 하여 출력해주기
avg = tot/3
print('총점 : {}, 평균:{}'.format(tot,avg))
```
